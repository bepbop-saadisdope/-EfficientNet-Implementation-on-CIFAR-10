---
## 1. Import Required Libraries

```python
# Import libraries

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

print(f"TensorFlow version: {tf.__version__}")

```

---
## 2. Download CIFAR-10 Dataset

```python
# Download CIFAR-10 dataset (automatic download)
print("Downloading CIFAR-10 dataset...")
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
print("Download complete!")

# Define class names
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']
```

---
## 3. Dataset Overview

```python
# Display dataset statistics
print("="*60)
print("CIFAR-10 DATASET STATISTICS")
print("="*60)
print(f"Training images: {x_train.shape[0]:,}")
print(f"Test images: {x_test.shape[0]:,}")
print(f"Total images: {x_train.shape[0] + x_test.shape[0]:,}")
print(f"\nImage dimensions: {x_train.shape[1]} × {x_train.shape[2]} pixels")
print(f"Color channels: {x_train.shape[3]} (RGB)")
print(f"Number of classes: {len(class_names)}")
print(f"\nPixel value range: {x_train.min()} to {x_train.max()}")
print(f"Data type: {x_train.dtype}")
print(f"\nTraining set shape: {x_train.shape}")
print(f"Training labels shape: {y_train.shape}")
print(f"Test set shape: {x_test.shape}")
print(f"Test labels shape: {y_test.shape}")
```

---
## 4. Class Distribution Analysis

```python
# Analyze class distribution
print("="*60)
print("CLASS DISTRIBUTION")
print("="*60)

# Training set distribution
unique_train, counts_train = np.unique(y_train, return_counts=True)
print("\nTraining Set:")
for i, (cls, count) in enumerate(zip(class_names, counts_train)):
    percentage = (count / len(y_train)) * 100
    print(f"{cls:12s}: {count:5d} images ({percentage:.1f}%)")

# Test set distribution
unique_test, counts_test = np.unique(y_test, return_counts=True)
print("\nTest Set:")
for i, (cls, count) in enumerate(zip(class_names, counts_test)):
    percentage = (count / len(y_test)) * 100
    print(f"{cls:12s}: {count:5d} images ({percentage:.1f}%)")
```

---
## 5. Data Preprocessing

```python
# Normalize pixel values to [0, 1]
print("Preprocessing data...")
print("\n1. Normalizing pixel values to [0, 1]...")

x_train_normalized = x_train.astype('float32') / 255.0
x_test_normalized = x_test.astype('float32') / 255.0

print(f"   Original range: [{x_train.min()}, {x_train.max()}]")
print(f"   Normalized range: [{x_train_normalized.min()}, {x_train_normalized.max()}]")
```

```python
# Convert labels to one-hot encoding
print("\n2. Converting labels to one-hot encoding...")

y_train_onehot = tf.keras.utils.to_categorical(y_train, 10)
y_test_onehot = tf.keras.utils.to_categorical(y_test, 10)

print(f"   Original label shape: {y_train.shape}")
print(f"   One-hot label shape: {y_train_onehot.shape}")
print(f"\n   Example - Original label: {y_train[0][0]}")
print(f"   Example - One-hot label: {y_train_onehot[0]}")
```

```python
# Summary of preprocessed data
print("\n" + "="*60)
print("PREPROCESSED DATA SUMMARY")
print("="*60)
print(f"Training images (normalized): {x_train_normalized.shape}")
print(f"Training labels (one-hot): {y_train_onehot.shape}")
print(f"Test images (normalized): {x_test_normalized.shape}")
print(f"Test labels (one-hot): {y_test_onehot.shape}")
print("\n Preprocessing complete!")
```

---
## 6. Save Preprocessed Data

```python
# Save preprocessed data to disk
print("Saving preprocessed data...")

np.save('x_train_normalized.npy', x_train_normalized)
np.save('x_test_normalized.npy', x_test_normalized)
np.save('y_train_onehot.npy', y_train_onehot)
np.save('y_test_onehot.npy', y_test_onehot)

# Also save original data for reference
np.save('x_train_original.npy', x_train)
np.save('x_test_original.npy', x_test)
np.save('y_train_original.npy', y_train)
np.save('y_test_original.npy', y_test)

print(" Preprocessed data saved!")
print("\nSaved files:")
print("  - x_train_normalized.npy")
print("  - x_test_normalized.npy")
print("  - y_train_onehot.npy")
print("  - y_test_onehot.npy")
print("  - x_train_original.npy")
print("  - x_test_original.npy")
print("  - y_train_original.npy")
print("  - y_test_original.npy")
```
